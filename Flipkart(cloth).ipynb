```python
def find_last_nos_in_a_file(enter_system_path):
    max_pages_in_file = 0
    for path,folder,file in os.walk(enter_system_path):
        for l in file:
            if int(re.findall('\d+',l)[0]) > max_pages_in_file:
                max_pages_in_file = int(re.findall('\d+',l)[0])
    return max_pages_in_file+1
```
