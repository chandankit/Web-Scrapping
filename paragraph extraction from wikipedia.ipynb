# store all image from wikipedia

```python
import urllib
```

```python
url = 'https://en.wikipedia.org/wiki/Mahatma_Gandhi'
list1 = []
direct_image=soup('img')

# print(direct_image)

for i in direct_image:
    s = i.get('src')
#     print('=====',s)
    print('======',s.find('Gandhi'))
    
    if s.find('Gandhi') > 0:
        if re.search('^http',s) != None:
            print('Starts with http')
        else:
            print(s)
            list1.append('http:'+ s)
print('^^^^^^^^^',list1)
c=0             
for l2 in list1:
    c=c+1                 
    url=l2
    fimg=open('C:\\Users\\Ankit\\ANKIT\\Web Scrapping Practice\\wiki\\'+'Gandhiji_'+str(c)+'.jpg','wb')   #wb=write binary becoz images are binary files
    data=urllib.request.urlopen(l2)
    buffersize=10000#first read and then write so we r fixing size
    info=data.read(buffersize)
    while info:
        fimg.write(info)
        info=data.read(buffersize)
    fimg.close()
```
